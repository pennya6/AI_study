```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
```

```python
traing_data=np.array([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
target_data=np.array([[0.],[0.],[0.],[1.]])
```

```python
inputs=keras.Input(shape=(2))
outputs=layers.Dense(1,activation='sigmoid')(inputs)
model=keras.Model(inputs=inputs,outputs=outputs,name='and_operator')
model.compile(loss='mse',optimizer='adam',metrics=['binary_accuracy'])
model.fit(traing_data,target_data,epochs=1000,verbose=2)
```

```python
model.summary()
```

```python
traing_data2=np.array([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
target_data2=np.array([[1.],[0.],[0.],[1.]])
```

```python
inputs=keras.Input(shape=(2))
x=layers.Dense(16,activation='sigmoid')(inputs)
outputs=layers.Dense(1,activation='sigmoid')(x)
model=keras.Model(inputs=inputs,outputs=outputs,name='xor_operator')
model.compile(loss='mse',optimizer='adam',metrics=['binary_accuracy'])
model.fit(traing_data2,target_data2,epochs=1000,verbose=1)
```

```python
model.summary()
```
